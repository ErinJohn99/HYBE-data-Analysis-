Cumulative return of 1.0: The investment value has not changed (0% gain or loss).

Cumulative return of 0.6: The investment has decreased by 40%.

```python
import numpy as np

# Calculate 5% value at risk
var_5 = np.percentile(data_cleaned['Daily Return'].dropna(), 5)
print(f"Value at Risk (5% confidence level): {var_5}")
```

The result of the calculation is: -0.04648875194731648, or about -4.65%.
This means that, with 95% confidence, the worst expected loss over a given period (one day return) will not exceed 4.65%.
In other words, there is only a 5% chance that the loss will be worse than 4.65% in a single day, based on the historical daily returns of the asset.

or in other words
This is your downside risk. It means that, on a bad day, you can expect the asset's value to decrease by as much as 4.65% with 95% confidence. Any larger loss has a 5% chance of occurring based on historical returns.

```python
import numpy as np


daily_var_5 = -0.04648875194731648  # 5% VaR
daily_var_10 = -0.03377726763742328  # 10% VaR

# Number of trading days in 6 months (assuming 21 trading days per month)
T_6_months = 21 * 6  # 6 months of trading days
# Number of trading days in 1 year
T_1_year = 21 * 12  # 1 year of trading days

# Calculate VaR for 6 months and 1 year
var_5_6_months = daily_var_5 * np.sqrt(T_6_months)
var_10_6_months = daily_var_10 * np.sqrt(T_6_months)

var_5_1_year = daily_var_5 * np.sqrt(T_1_year)
var_10_1_year = daily_var_10 * np.sqrt(T_1_year)

(var_5_6_months, var_10_6_months, var_5_1_year, var_10_1_year)
```

For the 5% confidence level:
6-month VaR: -52.18% (approximately)
1-year VaR: -73.80% (approximately)
This means that, with 95% confidence, the worst expected loss over 6 months could be up to 52.18%, and over 1 year it could be as much as 73.80%.

For the 10% confidence level:
6-month VaR: -37.91% (approximately)
1-year VaR: -53.62% (approximately)

```python
import numpy as np



# Calculate the 95th and 90th percentiles for the best-case daily returns (Value at Gain)
vag_95_daily = np.percentile(data_cleaned['Daily Return'].dropna(), 95)
vag_90_daily = np.percentile(data_cleaned['Daily Return'].dropna(), 90)

# Number of trading days in 6 months (assuming 21 trading days per month)
T_6_months = 21 * 6  # 6 months of trading days
# Number of trading days in 1 year
T_1_year = 21 * 12  # 1 year of trading days

# Calculate VaG for 6 months and 1 year
vag_95_6_months = vag_95_daily * np.sqrt(T_6_months)
vag_90_6_months = vag_90_daily * np.sqrt(T_6_months)

vag_95_1_year = vag_95_daily * np.sqrt(T_1_year)
vag_90_1_year = vag_90_daily * np.sqrt(T_1_year)

# Print the results
print(f"Value at Gain (95% confidence level, 6 months): {vag_95_6_months}")
print(f"Value at Gain (90% confidence level, 6 months): {vag_90_6_months}")
print(f"Value at Gain (95% confidence level, 1 year): {vag_95_1_year}")
print(f"Value at Gain (90% confidence level, 1 year): {vag_90_1_year}")
```

```python
window = 20
data_cleaned['MA20'] = data_cleaned['Close'].rolling(window=window).mean()
data_cleaned['Upper Band'] = data_cleaned['MA20'] + 2*data_cleaned['Close'].rolling(window=window).std()
data_cleaned['Lower Band'] = data_cleaned['MA20'] - 2*data_cleaned['Close'].rolling(window=window).std()

plt.figure(figsize=(12, 6))
plt.plot(data_cleaned['Date'], data_cleaned['Close'], label='Close Price', color='blue')
plt.plot(data_cleaned['Date'], data_cleaned['MA20'], label='20-day MA', color='orange')
plt.plot(data_cleaned['Date'], data_cleaned['Upper Band'], label='Upper Band', color='green')
plt.plot(data_cleaned['Date'], data_cleaned['Lower Band'], label='Lower Band', color='red')
plt.fill_between(data_cleaned['Date'], data_cleaned['Lower Band'], data_cleaned['Upper Band'], alpha=0.1)
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Bollinger Bands for HYBE Shares')
plt.legend()
plt.show()
```


Bollinger Bands are a type of volatility indicator that consists of three lines:

Middle Band (Yellow line): This is the 20-day simple moving average (SMA) of the stock's closing price. It provides a smoothed version of the stock's trend over 20 days.
Upper Band (Green line): This is the middle band plus two standard deviations of the closing price. It indicates the upper range within which the stock is expected to move, based on recent volatility.
Lower Band (Red line): This is the middle band minus two standard deviations of the closing price. It indicates the lower range within which the stock is expected to move.

Bollinger Bands Tightening and Widening (Volatility)

The distance between the upper and lower bands represents market volatility.
Widening of the bands (such as in 2021-07 and early 2022) indicates increased volatility, meaning that the price is moving more dramatically either up or down.
Narrowing of the bands (such as around mid-2023) indicates reduced volatility, meaning that the price is stabilizing or consolidating.

Price Touching the Upper or Lower Band

Price near the upper band (green line) typically indicates that the stock may be overbought and due for a pullback.
For example, around mid-2021, the price touches the upper band several times, followed by a correction.
Price near the lower band (red line) often indicates that the stock may be oversold and could experience a rebound.
This is visible in early 2022 when the price neared the lower band before bouncing back upward

Price Moving Between the Bands

The price typically moves within the bands, bouncing between the upper and lower bands. When it breaks above or below the bands, it can indicate strong momentum or market overreaction.
In the beginning of 2022, the price briefly dipped below the lower band, followed by a rapid reversal upwards. This is a classic Bollinger Band signal indicating that the market was temporarily oversold.

Mean Reversion

Bollinger Bands also suggest the concept of mean reversion, where the price tends to return to the middle band (the 20-day moving average) after touching the upper or lower bands.
For example, after hitting the upper band in mid-2021, the price reverted back to the middle band before stabilizing.

```python
delta = data_cleaned['Close'].diff(1)
gain = delta.where(delta > 0, 0)
loss = -delta.where(delta < 0, 0)
avg_gain = gain.rolling(window=14).mean()
avg_loss = loss.rolling(window=14).mean()
rs = avg_gain / avg_loss
data_cleaned['RSI'] = 100 - (100 / (1 + rs))

plt.figure(figsize=(12, 6))
plt.plot(data_cleaned['Date'], data_cleaned['RSI'], label='RSI', color='purple')
plt.axhline(70, linestyle='--', color='red')
plt.axhline(30, linestyle='--', color='green')
plt.xlabel('Date')
plt.ylabel('RSI Value')
plt.title('Relative Strength Index (RSI)')
plt.show()
```

RSI Line (Purple Line):

The purple line represents the RSI value over time, ranging from 0 to 100. It fluctuates as the stock price changes.
Overbought Threshold (Red Line at 70):

The horizontal red dashed line at RSI = 70 indicates the overbought threshold. When the RSI rises above 70, it suggests that the stock may be overbought and could be due for a correction or pullback.
In the chart, the RSI exceeds 70 at several points, such as in early 2022 and mid-2023, signaling potential overbought conditions.
Oversold Threshold (Green Line at 30):

The horizontal green dashed line at RSI = 30 represents the oversold threshold. When the RSI drops below 30, it suggests that the stock may be oversold, indicating a potential buying opportunity as the stock may rebound.
In the chart, the RSI falls below 30 during early 2021 and mid-2022, signaling potential oversold conditions.

Overbought Periods:
The RSI exceeds 70 several times throughout the period shown, notably around mid-2021 and early 2023. These instances suggest that HYBE shares were potentially overbought, which often leads to a price correction. After some of these peaks, the stock did experience a price decline.
Oversold Periods:
The RSI dips below 30 in early 2021 and mid-2022, signaling that the stock was potentially oversold. Following these periods, we see some price recoveries, which is typical after an RSI signals oversold conditions.
